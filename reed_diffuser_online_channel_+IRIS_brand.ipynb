```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Sample Data - Replace with actual data if available
data = {
    "Platform": ["Amazon", "Flipkart", "Tata CLiQ", "D2C Websites", "Retail Stores"],
    "Sales (%)": [35, 25, 15, 15, 10],
    "Discounts (%)": [10, 15, 12, 5, 8],
    "Festive Sales Increase (%)": [50, 45, 40, 35, 30]
}

fragrance_data = {
    "Fragrance": ["Lavender", "Sandalwood", "Jasmine", "Lemongrass", "Orange Blossom"],
    "Popularity (%)": [30, 25, 20, 15, 10]
}

price_data = {
    "Price Range (INR)": ["<500", "500-1000", "1000-2000", "2000-4000", ">4000"],
    "Market Share (%)": [10, 20, 35, 25, 10]
}

brand_sales_data = {
    "Brand": ["IRIS", "EKAM", "Song of India", "Westside Home", "Soulflower"],
    "Market Share (%)": [30, 25, 20, 15, 10]
}

regional_sales_data = {
    "Region": ["Metro Cities", "Tier 1 Cities", "Tier 2 Cities", "Tier 3 Cities"],
    "Sales (%)": [40, 30, 20, 10]
}

# Convert to DataFrames
df_sales = pd.DataFrame(data)
df_fragrance = pd.DataFrame(fragrance_data)
df_price = pd.DataFrame(price_data)
df_brands = pd.DataFrame(brand_sales_data)
df_regions = pd.DataFrame(regional_sales_data)

# ----------------------------- Graph 1: Market Share (Platform-wise Sales) -----------------------------
plt.figure(figsize=(8, 5))
plt.bar(df_sales["Platform"], df_sales["Sales (%)"], color='blue')
plt.xlabel("Sales Platforms")
plt.ylabel("Sales Share (%)")
plt.title("Platform-wise Market Share")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# ----------------------------- Graph 2: Price Segmentation -----------------------------
plt.figure(figsize=(8, 5))
plt.bar(df_price["Price Range (INR)"], df_price["Market Share (%)"], color='green')
plt.xlabel("Price Range (INR)")
plt.ylabel("Market Share (%)")
plt.title("Price-wise Market Share Distribution")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# ----------------------------- Graph 3: Sales Growth (Festive Sales Increase) -----------------------------
plt.figure(figsize=(8, 5))
plt.plot(df_sales["Platform"], df_sales["Festive Sales Increase (%)"], marker='o', linestyle='-', color='red')
plt.xlabel("Platforms")
plt.ylabel("Sales Increase During Festivals (%)")
plt.title("Impact of Festivals on Sales Growth")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.show()

# ----------------------------- Graph 4: Discounts Across Platforms -----------------------------
plt.figure(figsize=(8, 5))
plt.bar(df_sales["Platform"], df_sales["Discounts (%)"], color='purple')
plt.xlabel("Platforms")
plt.ylabel("Average Discount (%)")
plt.title("Discount Strategies by Platforms")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# ----------------------------- Graph 5: Fragrance Preferences -----------------------------
plt.figure(figsize=(6, 6))
plt.pie(df_fragrance["Popularity (%)"], labels=df_fragrance["Fragrance"], autopct='%1.1f%%', startangle=140)
plt.title("Fragrance Preferences in India")
plt.show()

# ----------------------------- Graph 6: Competitive Market Share (Brand-wise) -----------------------------
plt.figure(figsize=(8, 5))
plt.bar(df_brands["Brand"], df_brands["Market Share (%)"], color='orange')
plt.xlabel("Brands")
plt.ylabel("Market Share (%)")
plt.title("Top Reed Diffuser Brands in India")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# ----------------------------- Graph 7: Platform-wise Sales Trend (Line Chart) -----------------------------
plt.figure(figsize=(8, 5))
plt.plot(df_sales["Platform"], df_sales["Sales (%)"], marker='o', linestyle='-', color='blue')
plt.xlabel("Sales Platforms")
plt.ylabel("Sales Share (%)")
plt.title("Platform-wise Sales Trend")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.show()

# ----------------------------- Graph 8: Regional Sales Distribution -----------------------------
plt.figure(figsize=(8, 5))
plt.bar(df_regions["Region"], df_regions["Sales (%)"], color='brown')
plt.xlabel("Regions")
plt.ylabel("Sales Share (%)")
plt.title("Regional Sales Distribution of Reed Diffusers")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# ----------------------------- Graph 9: Scatter Plot (Price vs Market Share) -----------------------------
plt.figure(figsize=(8, 5))
plt.scatter(df_price["Market Share (%)"], df_price.index, color='cyan', edgecolor='black')
plt.xlabel("Market Share (%)")
plt.ylabel("Price Category")
plt.title("Price vs Market Share")
plt.grid(alpha=0.7)
plt.show()

# ----------------------------- Graph 10: Heatmap (Platform Performance) -----------------------------
performance_data = df_sales.set_index("Platform").drop(columns=["Sales (%)"])  # Drop sales for heatmap focus
plt.figure(figsize=(8, 5))
sns.heatmap(performance_data, cmap="coolwarm", annot=True, fmt=".1f", linewidths=0.5)
plt.title("Platform Performance Heatmap (Discounts & Festival Impact)")
plt.xlabel("Metrics")
plt.ylabel("Platforms")
plt.show()
```

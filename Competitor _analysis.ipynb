```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#  1 CREATE SYNTHETIC DATASET (Brand, Product Type, Price, Sales, Season, Festival, Discount, Fragrance)
np.random.seed(42)

brands = ["Godrej aer", "Ambi Pur", "Air Wick", "Odonil", "Glade", "Mangalam CamPure"]
product_types = ["Room Spray", "Car Freshener", "Automatic Spray", "Bathroom Freshener", "Gel", "Camphor Purifier"]
seasons = ["Winter", "Summer", "Monsoon", "Autumn"]
festivals = ["Diwali", "Christmas", "Holi", "None"]
discounts = ["No Discount", "10% Off", "20% Off", "Buy 1 Get 1"]
fragrances = ["Lavender", "Jasmine", "Lemon", "Rose", "Sandalwood", "Ocean Breeze"]

data = []

for i in range(500):  # Generate 500 rows of sales data
    brand = np.random.choice(brands)
    product = np.random.choice(product_types)
    season = np.random.choice(seasons)
    festival = np.random.choice(festivals, p=[0.2, 0.1, 0.15, 0.55])  # 55% of sales happen in non-festival periods
    discount = np.random.choice(discounts, p=[0.5, 0.2, 0.2, 0.1])  # 50% sales happen without discounts
    fragrance = np.random.choice(fragrances)
    price = np.random.randint(100, 500)  # Price range
    sales = np.random.randint(50, 500)  # Units sold

    data.append([brand, product, price, season, festival, discount, fragrance, sales])

# Convert to DataFrame
df = pd.DataFrame(data, columns=["Brand", "Product Type", "Price", "Season", "Festival", "Discount", "Fragrance", "Sales"])

#  2 SHOW DATA OVERVIEW
print("\n Dataset Sample:\n", df.head())

#  3 SEASONAL SALES ANALYSIS
plt.figure(figsize=(10, 6))
sns.boxplot(x="Season", y="Sales", data=df, palette="coolwarm")
plt.title("Seasonal Sales Trend")
plt.xlabel("Season")
plt.ylabel("Units Sold")
plt.show()

#  4 FESTIVAL SALES ANALYSIS
plt.figure(figsize=(10, 6))
sns.barplot(x="Festival", y="Sales", data=df, palette="viridis", estimator=np.sum)
plt.title("Sales During Festivals")
plt.xlabel("Festival")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()

#  5 DISCOUNT EFFECT ON SALES
plt.figure(figsize=(10, 6))
sns.boxplot(x="Discount", y="Sales", data=df, palette="coolwarm")
plt.title("Impact of Discounts on Sales")
plt.xlabel("Discount")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.show()

#  6 BEST SELLING FRAGRANCES
plt.figure(figsize=(10, 6))
sns.barplot(x=df["Fragrance"].value_counts().index, y=df["Fragrance"].value_counts().values, palette="viridis")
plt.title("Top-Selling Fragrances")
plt.xlabel("Fragrance")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()

#  7 BRAND-WISE SALES TREND
plt.figure(figsize=(10, 6))
sns.barplot(x="Brand", y="Sales", data=df, palette="magma", estimator=np.sum)
plt.title("Brand-Wise Total Sales")
plt.xlabel("Brand")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()

#  8 PRODUCT-WISE PRICE TREND
plt.figure(figsize=(10, 6))
sns.boxplot(x="Product Type", y="Price", data=df, palette="Set3")
plt.title("Price Distribution by Product Type")
plt.xlabel("Product Type")
plt.ylabel("Price")
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#  1 CREATE SYNTHETIC DATASET (Brand, Product Type, Price, Sales, Season, Festival, Discount, Fragrance, State, Distribution Channel)
np.random.seed(42)

brands = ["Godrej aer", "Ambi Pur", "Air Wick", "Odonil", "Glade", "Mangalam CamPure"]
product_types = ["Room Spray", "Car Freshener", "Automatic Spray", "Bathroom Freshener", "Gel", "Camphor Purifier"]
states = ["Maharashtra", "Gujarat", "Rajasthan", "Delhi", "Uttar Pradesh", "Punjab",
          "Tamil Nadu", "Kerala", "Karnataka", "Andhra Pradesh", "West Bengal", "Odisha", "Bihar"]
distribution_channels = ["General Trade", "Modern Trade", "E-commerce", "Quick Commerce"]

data = []

for i in range(1000):  # Generate 1000 rows of sales data
    brand = np.random.choice(brands)
    product = np.random.choice(product_types)
    state = np.random.choice(states)
    distribution_channel = np.random.choice(distribution_channels, p=[0.4, 0.3, 0.2, 0.1])  # Weighted distribution
    price = np.random.randint(100, 500)  # Price in 
    sales = np.random.randint(50, 500)  # Sales in units

    data.append([brand, product, state, distribution_channel, price, sales])

# Convert to DataFrame
df = pd.DataFrame(data, columns=["Brand", "Product Type", "State", "Distribution Channel", "Price ()", "Sales (Units)"])

#  2 DISPLAY DATA SAMPLE
print("\n Dataset Sample:\n", df.head())

#  3 DISTRIBUTION CHANNEL DOMINANCE BY BRAND
plt.figure(figsize=(12, 6))
sns.countplot(y=df["Brand"], hue=df["Distribution Channel"], data=df, palette="coolwarm")
plt.title("Distribution Channel Dominance by Brand", fontsize=14, fontweight="bold")
plt.xlabel("Count of Sales Entries")
plt.ylabel("Brand")
plt.legend(title="Distribution Channel")
plt.show()

#  4 DISTRIBUTION CHANNEL DOMINANCE BY PRODUCT TYPE
plt.figure(figsize=(12, 6))
sns.countplot(y=df["Product Type"], hue=df["Distribution Channel"], data=df, palette="viridis")
plt.title("Distribution Channel Dominance by Product Type", fontsize=14, fontweight="bold")
plt.xlabel("Count of Sales Entries")
plt.ylabel("Product Type")
plt.legend(title="Distribution Channel")
plt.show()

#  5 DISTRIBUTION CHANNEL DOMINANCE BY STATE
plt.figure(figsize=(12, 6))
sns.countplot(y=df["State"], hue=df["Distribution Channel"], data=df, palette="magma")
plt.title("Distribution Channel Dominance by State", fontsize=14, fontweight="bold")
plt.xlabel("Count of Sales Entries")
plt.ylabel("State")
plt.legend(title="Distribution Channel")
plt.show()

#  6 DISTRIBUTION CHANNEL SALES SHARE ANALYSIS (STACKED BAR)
pivot_df = df.groupby(["Distribution Channel", "Brand"])["Sales (Units)"].sum().unstack()

pivot_df.plot(kind="bar", stacked=True, figsize=(12, 6), colormap="coolwarm")
plt.title("Brand-wise Sales Across Distribution Channels", fontsize=14, fontweight="bold")
plt.xlabel("Distribution Channel", fontsize=12)
plt.ylabel("Total Sales (Units)", fontsize=12)
plt.legend(title="Brand")
plt.xticks(rotation=45)
plt.show()

#  7 DISTRIBUTION CHANNEL SALES SHARE ANALYSIS BY PRODUCT TYPE
pivot_df_product = df.groupby(["Distribution Channel", "Product Type"])["Sales (Units)"].sum().unstack()

pivot_df_product.plot(kind="bar", stacked=True, figsize=(12, 6), colormap="viridis")
plt.title("Product Type-wise Sales Across Distribution Channels", fontsize=14, fontweight="bold")
plt.xlabel("Distribution Channel", fontsize=12)
plt.ylabel("Total Sales (Units)", fontsize=12)
plt.legend(title="Product Type")
plt.xticks(rotation=45)
plt.show()

#  8 DISPLAY FINAL DATASET FOR DETAILED ANALYSIS
import ace_tools as tools

tools.display_dataframe_to_user(name="Brand, Product, State, and Distribution Analysis", dataframe=df)
```
